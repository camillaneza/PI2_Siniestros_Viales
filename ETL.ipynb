```python
ruta_homicidios_hechos = 'Data/homicidios.xlsx'
```

```python
df_homicidios_hechos = pd.read_excel(ruta_homicidios_hechos, sheet_name='HECHOS')
df_homicidios_hechos.head()
```

Se puede observar que los nombres de las columnas estan diferentes, ya que algunas estan en mayúsculas y otras en minúsculas. Se transforman todas a mayúsculas, y se cambia el '_' por espacio ' ':

```python
df_homicidios_hechos.columns = df_homicidios_hechos.columns.str.upper()
```

```python
df_homicidios_hechos.columns = df_homicidios_hechos.columns.str.replace('_', ' ')
```

También se reenombran algunas columnas, para comprender mejor a qué se refieren:

```python
df_homicidios_hechos.rename(columns={'N_VICTIMAS': 'NUMERO DE VICTIMAS', 'AAAA': 'AÑO', 'MM': 'MES', 'DD': 'DIA', 'HH': 'HORA SIN MINUTOS'}, inplace=True)

df_homicidios_hechos.columns
```

Se verifica el tipo de dato de cada columna y se analiza si hay valores nulos:

```python
Funciones.verificar_tipo_datos(df_homicidios_hechos)
```

Se puede observar que la columna 'Altura' posee 81,47% de valores nulos, la columna 'Cruce' 24,57%, y la columna 'Dirección Normalizada' 1,15%.

```python
df_nulos__homicidios_hechos = df_homicidios_hechos[['ALTURA', 'CRUCE', 'DIRECCIÓN NORMALIZADA']]
df_nulos__homicidios_hechos.head(10)
```

Se procede a eliminar la columna 'Altura', por poseer muchos valores nulos, y siendo que resulta inútil teniendo la columna 'Dirección Normalizada':

```python
df_homicidios_hechos = df_homicidios_hechos.drop('ALTURA', axis=1)
```

Se analiza la columna 'Cruce', la cual también poseía nulos:

```python
df_nulos_cruce__homicidios_hechos = df_homicidios_hechos[['CRUCE', 'DIRECCIÓN NORMALIZADA']]
df_nulos_cruce__homicidios_hechos.head(40)
```

Se puede deducir que donde hay valores nulos, el accidente no se produjo en una esquina o en el cruce propiamente dicho; por lo que se cambia los valores nulos por "No fue en cruce de calles", y los demás valores por 'Fue en cruce de calles':

```python
df_homicidios_hechos['CRUCE'] = np.where(df_homicidios_hechos['CRUCE'].notnull(), 'FUE EN CRUCE DE CALLES', 'NO FUE EN CRUCE DE CALLES')
```

```python
df_homicidios_hechos[['CRUCE', 'DIRECCIÓN NORMALIZADA']].head(20)
```

Se observa la última columna que posee valores nulos, 'Dirección Normalizada':

```python
df_homicidios_hechos[df_homicidios_hechos['DIRECCIÓN NORMALIZADA'].isnull()][['DIRECCIÓN NORMALIZADA', 'LUGAR DEL HECHO', 'TIPO DE CALLE']]
```

Se observa que los valores faltantes son de accidentes en Autopista, excepto uno de ellos que fue en calle pero no se tienen datos del lugar del hecho (SD). Se cambian los nulos por 'SD' haciendo alusión a 'Sin Datos' y siguiendo la estructura del dataset:

```python
df_homicidios_hechos['DIRECCIÓN NORMALIZADA'].fillna("SD", inplace=True)
```

```python
df_homicidios_hechos[['DIRECCIÓN NORMALIZADA']].head(40)
```

Las columnas 'HORA', 'HORA SIN MINUTOS', y 'CALLE' poseen varios tipos de datos. 

```python
df_homicidios_hechos['HORA'].apply(type).value_counts()
```

```python
df_homicidios_hechos['HORA SIN MINUTOS'].apply(type).value_counts()
```

```python
df_homicidios_hechos['CALLE'].apply(type).value_counts()
```

Se decide cambiar el tipo de datos de la columna 'HORA' a datetime.time ya que representa solo la parte de la hora del día, sin incluir la fecha (como lo hace datetime.datetime, y en este caso no es necesario); el tipo de datos de la columna 'HORA SIN MINUTOS' a integer; y el tipo de datos de la columna 'CALLE' a string:

```python
df_homicidios_hechos['HORA'] = df_homicidios_hechos['HORA'].apply(Funciones.convertir_a_time)
```

```python
df_homicidios_hechos['HORA'].apply(type).value_counts()
```

```python
df_homicidios_hechos['HORA'].head(30)
```

Tanto la columna 'Hora' como la columna 'Hora sin minutos' poseen 'SD', es decir que no cuentan con datos. Se decide aplicar la moda

```python
moda_de_hora = df_homicidios_hechos['HORA'].mode().iloc[0]
moda_de_hora
```

```python
df_homicidios_hechos['HORA'].fillna(moda_de_hora, inplace=True)
```

```python
df_homicidios_hechos['HORA'].apply(type).value_counts()
```

```python
df_homicidios_hechos['HORA']
```

```python
df_homicidios_hechos['HORA SIN MINUTOS'] = df_homicidios_hechos['HORA SIN MINUTOS'].apply(lambda x: int(moda_de_hora.hour) if x == "SD" else x)
```

```python
df_homicidios_hechos['HORA SIN MINUTOS'].apply(type).value_counts()
```

```python
df_homicidios_hechos['HORA SIN MINUTOS']
```
